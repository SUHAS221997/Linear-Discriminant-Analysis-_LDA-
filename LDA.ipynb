**Difference between PCA and LDA**


LDA is supervised and PCA is unsupervised


LDA is used for maximizing the separation between two or more classes, whereas the PCA is used for maximizing the variation

```python
#Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

Load Digits

```python
dataset=load_digits()
```

```python
dataset.keys()
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df.isnull().sum()
```

```python
#Adding target to Dataset
df["target"]=dataset.target
```

```python
# Feature of the images
x_df=df.drop("target",axis=1)

#dependent feature
y_df=df["target"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

```python
#Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df["target"].nunique()
```

```python
lda=LinearDiscriminantAnalysis(n_components=min(10-1,64))
lda
```

```python
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
lda.coef_
```

```python
# The explained variance (how much variance is explained by each component)
explained_variance_ratio=lda.explained_variance_ratio_
explained_variance_ratio
```

```python
pd.DataFrame(explained_variance_ratio)
```

```python
pd.DataFrame(lda.coef_)
```

```python
sum(explained_variance_ratio)
```

```python
x_train.shape,x_test.shape
```

```python
# import Random Forest classifier
from sklearn.ensemble import RandomForestClassifier

# instantiate the classifier
clf = RandomForestClassifier(random_state=0,n_estimators= 100)

# fit the classifier with x and y data
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

cm
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
